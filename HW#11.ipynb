Problem 12.5 #7

From the model, we have the following: 
$\Delta x = (x - xy - \frac{3}{4}y)\Delta t$ \\
$\Delta y = (xy - y - \frac{3}{4}y)\Delta t$ \\
with $\Delta t$ = 1, x(0) = $\frac{1}{2}$, y(0)=1

```python
delta_t = 1 
x = 1/2
y=1
print('t:',0, 'x:',x, 'y:',y) 
for t in range(1,5):
  delta_x = x - x*y - (3/4)*y 
  delta_y = x*y - y - (3/4)*x
  x += delta_x*delta_t 
  y += delta_y*delta_t
  print('t:',t, 'x:',x, 'y:',y)
```

$\frac{d^2Q}{dx^2} = 3.78 - 0.54y < 0 for y =14$ \\

Gradient Ascent method. Using initial guess (x_0,y_0) = (1,1) and letting $\delta$ = 1.1 , $\lambda$ =0.01. Within 50 iterations, it can be seen that we converge to the correct values computed earlier (18,14) 

```python
x=1
y=1 
delta = 1.1
learning_rate = .01

dx = 1.08*x - .06*(x**2)
dy = 3.78*y - .27*(y**2)

n = 50

print('i:',0,'x:',x, 'y:',y) 
for i in range(1,n):
  x += dx*learning_rate 
  y += dy*learning_rate
  learning_rate *= delta

  dx = 1.08*x - .06*(x**2)
  dy = 3.78*y - .27*(y**2) 

  print('i:',i,'x:',x, 'y:',y)
```

Section 13.3 #3 

Distance function from $x^2 + y^2 - z^2 = 1$ to the origin: $(x-0)^2 + (y-0)^2+(z-0)^2 = x^2 + y^2 +z^2 $\\
We want to minimize \\
f(x,y,z) = $ x^2 + y^2 +z^2 $ \\
st.  $x^2 + y^2 - z^2 = 1$ \\
Thus we formulate the model as \\
L(x,y,z,$\lambda$) = f(x,y,z) + $\lambda$( $x^2 + y^2 - z^2 - 1)$ \\
= $ x^2 + y^2 +z^2 + \lambda x^2 + \lambda y^2 - \lambda z^2 - \lambda $ \\


$\frac{dL}{dx} = 2x + 2\lambda$ x = 0 \\
$\frac{dL}{dy} = 2y + 2\lambda$ y = 0 \\
$\frac{dL}{dz} = 2z - 2\lambda$ z = 0 \\
$\frac{dL}{d \lambda } = x^2 + y^2 - z^2 - 1 = 0$ \\

Using Wolfram Alpha to solve this system of equations, we get two solutions, one of which contains complex numbers which cannot be true for distances. The other solution is \\
$\lambda = -1, y = +- \sqrt{1-x^2}$, z = 0 \\
Thus the minimum distance is f(x,y,z) = $ x^2 + y^2 +z^2 = 1 $
